## Plot function

```python
def plot_combined_figure(x, y1, y2):
    """
    Plot both scatter and bar plots side by side.

    :param x: List of x-axis values.
    :param y1: List of y-axis values for the first plot.
    :param y2: List of y-axis values for the second plot.
    """
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Plot scatter plot on the first subplot
    axs[0].plot(x, y1, marker='o', linestyle='None', color='blue', label='cosine')
    axs[0].plot(x, y2, marker='o', linestyle='None', color='red', label='tanimoto')
    mean_y1 = np.mean(y1)
    mean_y2 = np.mean(y2)
    axs[0].axhline(mean_y1, color='blue', linestyle='--', label=f'Mean (cosine): {mean_y1:.2f}')
    axs[0].axhline(mean_y2, color='red', linestyle='--', label=f'Mean (tanimoto): {mean_y2:.2f}')
    axs[0].set_xlabel('Complexe')
    axs[0].set_ylabel('Distance to the other compounds')
    axs[0].set_title("Distribution of the Distances")
    axs[0].legend()
    axs[0].grid(True)

    # Plot bar plot on the second subplot
    y_min = min(min(y1), min(y2))
    y_max = max(max(y1), max(y2))
    y_range = y_max - y_min
    interval_size = y_range / 10
    
    # Calculate histogram for y1 and y2
    interval_counts_y1, _ = np.histogram(y1, bins=10, range=(y_min, y_max))
    interval_counts_y2, _ = np.histogram(y2, bins=10, range=(y_min, y_max))
    
    bar_width = 0.35
    color_y1 = 'blue'
    color_y2 = 'red'
    axs[1].bar(np.arange(10) - bar_width/2, interval_counts_y1, width=bar_width, align='center', color=color_y1, label='cosine')
    axs[1].bar(np.arange(10) + bar_width/2, interval_counts_y2, width=bar_width, align='center', color=color_y2, label='tanimoto')
    interval_ranges = [(y_min + i * interval_size, y_min + (i + 1) * interval_size) for i in range(10)]
    axs[1].set_xticks(np.arange(10))
    axs[1].set_xticklabels([f"{interval[0]:.2f}-{interval[1]:.2f}" for interval in interval_ranges], rotation=45)
    axs[1].set_xlabel('Minimal Distance Intervals')
    axs[1].set_ylabel('Number of complexes having this minimal distance')
    axs[1].set_title('Distribution of Minimal Distances')
    axs[1].legend()
    axs[1].grid(True)

    plt.tight_layout()
    plt.show()
```

```python
def plot_combined_distances(distances1, distances2):
    
    # Ensure distances1 and distances2 are of the same length
    # When a complex has some NaN values in the descriptors (very rare but it can happen), then the 
    #distances list with descriptors doesn't count it.
    
    min_length = min(len(distances1), len(distances2))
    if len(distances1) > min_length:
        distances1 = random.sample(distances1, min_length)
    elif len(distances2) > min_length:
        distances2 = random.sample(distances2, min_length)
    
    complexes = [i for i in range(len(distances1))]
    plot_combined_figure(complexes, distances1, distances2)
```

```python
def get_minimal_distances(df, distance, desc = False): 
    
    """
    Returns a list of all the minimal distance between a finegrprint and the rest of the library.
    param : df: the library of generated compounds 
    param : distance : the distance used
    return : a list of floats (the distances)
    
    """
    
    minimal_distances = []
    
    for index, row in df.iterrows():
        if desc == True : 
            fingerprint = row['Descriptors']
            list = get_distances(df, fingerprint, distance, desc = True)
            if len(list) !=0 :
                minimal_distance = min(list) 
                minimal_distances.append(minimal_distance)
        else : 
            fingerprint = row['Fingerprint']
            minimal_distance = min(get_distances(df, fingerprint, distance))
            minimal_distances.append(minimal_distance)
        
    return minimal_distances
```

## Distance between synthesized complexes and dataset

```python
synth = pd.read_csv('synthesized_complexes.csv')
```

```python
synth['SMILES'] = synth.L1+'.'+synth.L2+'.'+synth.L3
```

```python
synth['mol'] = synth['SMILES'].apply(Chem.MolFromSmiles)
```

```python
synth['MOL1'] = synth.L1.apply(Chem.MolFromSmiles)
synth['MOL2'] = synth.L2.apply(Chem.MolFromSmiles)
synth['MOL3'] = synth.L3.apply(Chem.MolFromSmiles)
```

```python
fp_generator = CheMeleonFingerprint(device=None)

synth['CheMeleonFP_1'] = list(fp_generator(synth["MOL1"].tolist()))
synth['CheMeleonFP_2'] = list(fp_generator(synth["MOL2"].tolist()))
synth['CheMeleonFP_3'] = list(fp_generator(synth["MOL3"].tolist()))
```

```python
add_lists = lambda row: [sum(x) for x in zip(row['CheMeleonFP_1'], row['CheMeleonFP_2'], row['CheMeleonFP_3'])]
synth['Fingerprint'] = synth.apply(add_lists, axis=1)

#fp = StandardScaler().fit_transform(synth['Fingerprint'].tolist())
```

```python
fp = synth['Fingerprint'].tolist()
```

```python
fp = np.array(fp)
```

```python
fp = StandardScaler().fit_transform(fp)
```

```python
synth['Fingerprint_scaled'] = fp.tolist()
```

```python
synth['Fingerprint'] = synth['Fingerprint_scaled']
```

```python
def get_min_distances_synth(df_synth, dataset):
    minimal_distances_cosine = []
    minimal_distances_tanimoto = []
    for index, row in df_synth.iterrows():
        list_cosine = get_distances(dataset, row['Fingerprint'], cosine_distance)
        if len(list_cosine) !=0 :
            minimal_distances_cosine.append(min(list_cosine))
        list_tanimoto = get_distances(dataset, row['Fingerprint'], tanimoto_distance)
        if len(list_tanimoto) !=0 :
            minimal_distances_tanimoto.append(min(list_tanimoto))
    output_df = synth.copy()
    output_df['min_cosine_dist'] = minimal_distances_cosine
    output_df['min_tanimoto_dist'] = minimal_distances_tanimoto

    return output_df
```

```python
min_dist = get_min_distances_synth(synth, metals_chemeleon)
```

```python
import mols2grid 
subset=['min_cosine_dist', 'min_tanimoto_dist']
transform = {col: (lambda v, c=col: f"{c} : {v}") for col in subset}

mols2grid.display(min_dist, size=(300,300), transform=transform, subset=subset, useSVG=True)
```
