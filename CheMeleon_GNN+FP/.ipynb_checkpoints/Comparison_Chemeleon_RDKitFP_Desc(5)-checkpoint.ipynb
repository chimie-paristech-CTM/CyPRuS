```python
def plot_model_agreement(y_true, y_pred_desc, y_pred_comp, model_compared):

    agree = -np.abs(y_pred_desc- y_pred_comp)
    agree_norm = (agree - agree.min()) / (agree.max() - agree.min())
    print(agree_norm.min(), agree_norm.max())
    plt.figure(figsize=(7,7))
    plt.plot(y_true, y_true, linestyle="--", color='lightgray')
    #plt.scatter(y_true, y_pred_desc)
    plt.scatter(
        y_true,
        y_pred_desc,
        c=agree_norm, # color = agreement
        alpha = 0.8,
        s=20,
        cmap='rocket'
    )
    plt.xlabel("True pIC50", fontsize=14)
    plt.ylabel("Predicted pIC50", fontsize=14)
    plt.title(f"Correlation plot \n Agreement between {model_compared} \nand the RDKit descriptors", fontsize=16)

    cbar = plt.colorbar()
    cbar.set_label("Agreement", fontsize=12)

    plt.tight_layout()
    #plt.savefig(f'comparison_desc_{model_compared}.png', dpi=300)
    plt.show()
```

```python
plot_model_agreement(y_data, y_predictions_desc, y_predictions_rdkit, 'RDKit fingerprints')
```

```python
plot_model_agreement(y_data, y_predictions_desc, y_predictions_chemeleon, 'CheMeleon GNN')
```

```python
def hexbin_agreement_plot(x, y, y_ref, encoding):

  # agreement in [0,1]
  agreement = 1.0 - np.clip(np.abs(y- y_ref) / np.max(np.abs(y - y_ref)), 0, 1)

  fig, ax = plt.subplots(figsize=(7,7))

  # gridsize controls resolution (bigger = finer)
  hb = ax.hexbin(x, y_ref, C=agreement, reduce_C_function=np.mean,
              gridsize=20, cmap='rocket', mincnt=1, alpha=0.75)  # mincnt=1 hides empty bins

  # overlay scatter (optional — small alpha for speed/visibility)
  ax.scatter(x, y, c=agreement, cmap='rocket', s=20, edgecolor='none', alpha=0.8)

  ax.plot([ax.get_xlim()[0], ax.get_xlim()[1]],
      [ax.get_xlim()[0], ax.get_xlim()[1]],
      '--', color='k')

  cbar = fig.colorbar(hb, ax=ax)
  cbar.set_label('Local mean agreement', fontsize=12)
  ax.set_xlabel('True pIC50', fontsize=14)
  ax.set_ylabel('Predicted pIC50', fontsize=14)
  ax.set_title(f'Correlation plot \n {encoding} agreement with Descriptors', fontsize=16)
  plt.tight_layout()
  #plt.savefig(f'agreement_desc_{encoding}_hexbin.png', dpi=300)
  plt.show()
```

```python
hexbin_agreement_plot(y_data, y_predictions_rdkit, y_predictions_desc, 'RDKit FP')
```

```python
hexbin_agreement_plot(y_data, y_predictions_chemeleon, y_predictions_desc, 'CheMeleon GNN')
```

#### Conditional colouring with threshold

```python
def plot_model_agreement_threshold(y_true, y_pred_desc, y_pred_comp, model_compared, threshold):

    agree = -np.abs(y_pred_desc- y_pred_comp)
    agree_norm = (agree - agree.min()) / (agree.max() - agree.min())
    print(agree_norm.min(), agree_norm.max())

    y_true_gray = []
    y_pred_gray = []

    y_true_color = []
    y_pred_color = []
    
    for i in range(len(agree_norm)):
        
        if agree_norm[i] < threshold :
            y_true_gray.append(y_true[i])
            y_pred_gray.append(y_pred_desc[i])
        else:
            y_true_color.append(y_true[i])
            y_pred_color.append(y_pred_desc[i])
    
    plt.figure(figsize=(7,7))
    plt.plot(y_true, y_true, linestyle="--", color='black', zorder=1)
    #plt.scatter(y_true, y_pred_desc)
    plt.scatter(
        y_true_gray,
        y_pred_gray,
        c='lightgray',
        alpha = 0.8,
        s=20,
        zorder=2,
    )

    plt.scatter(
        y_true_color,
        y_pred_color,
        color='#E7ACFC',
        alpha = 1,
        s=20,
        zorder=3,
    )

    plt.xlabel("True pIC50", fontsize=14)
    plt.ylabel("Predicted pIC50", fontsize=14)
    plt.title(f"Correlation plot \n Agreement between {model_compared} \nand the RDKit descriptors", fontsize=16)

    plt.tight_layout()
    plt.savefig(f'comparison_desc_{model_compared}_scaffoldB_threshold_{threshold}.png', dpi=300)
    plt.show()
```

```python
plot_model_agreement_threshold(y_data, y_predictions_desc, y_predictions_chemeleon, 'CheMeleon GNN', 0.9)
```

```python
plot_model_agreement_threshold(y_data, y_predictions_desc, y_predictions_rdkit, 'RDKIT', 0.9)
```
