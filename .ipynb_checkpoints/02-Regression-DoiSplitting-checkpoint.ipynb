### Useful Functions

```python
def get_doi(fingerprint, df):
    """
    Get the paper's DOI for a fingerprint.

    :param fingerprint: a fingerprint for the metal complexe (sum of 3 ligand's fingerprints).
    :df: dataset used.
    :return: the corresponding doi
    """
    
    fingerprints_list = df['Fingerprint'].tolist()
    row_index = [i for i, fp in enumerate(fingerprints_list) if (fp == fingerprint)]
    doi = df.iloc[row_index]['DOI']

    return doi
```

```python
def doi_list(fingerprints, df, use_indices=False):
    """
    Computes dois for each smiles string and returns a mapping from dois to sets of smiles.

    :param mols: A list of smiles strings or RDKit molecules.
    :param use_indices: Whether to map to the smiles' index in all_smiles rather than mapping
    to the smiles string itself. This is necessary if there are duplicate smiles.
    :return: A dictionary mapping each unique doi to all fingerprints (or fingerprints indices) which have that doi.
    """
    dois = defaultdict(set)
    for i, fingerprint in tqdm(enumerate(fingerprints), total=len(fingerprints)):
        doi = get_doi(fingerprint, df)
        if use_indices:
            # Convert the DOI Series to a string for use as a dictionary key
            doi_str = str(doi.values[0])
            dois[doi_str].add(i)
        else:
            # Convert the DOI Series to a string for use as a dictionary key
            doi_str = str(doi.values[0])
            dois[doi_str].add(fingerprint)

    return dois
```

```python
# modify this function from Chemprop to just return the indices
def doi_split(df,
                   sizes=(0.8, 0.1, 0.1),
                   balanced=False,
                   seed=0,
                   ):
    """
    Split a dataset by doi so that no molecules sharing a doi are in the same split.

    :param data: A MoleculeDataset or ReactionDataset.
    
    :param fingerprint: a list of fingerprints
    
    :param sizes: A length-3 tuple with the proportions of data in the
    train, validation, and test sets.
    :param balanced: Try to balance sizes of dois in each set, rather than just putting smallest in test set.
    :param seed: Seed for shuffling when doing balanced splitting.
    :return: A tuple containing the train, validation, and test splits of the data.
    """
    assert sum(sizes) == 1

    fingerprints = df['Fingerprint'].tolist()
    
    # Split
    train_size, val_size, test_size = sizes[0] * len(fingerprints), sizes[1] * len(fingerprints), sizes[2] * len(fingerprints)
    train, val, test = [], [], []
    train_doi_count, val_doi_count, test_doi_count = 0, 0, 0
    
    doi_to_indices = doi_list(fingerprints, df, use_indices=True)
    
    
    if balanced:  # Put stuff that's bigger than half the val/test size into train, rest just order randomly
        index_sets = list(doi_to_indices.values())
        big_index_sets = []
        small_index_sets = []
        for index_set in index_sets:
            if len(index_set) > val_size / 2 or len(index_set) > test_size / 2:
                big_index_sets.append(index_set)
            else:
                small_index_sets.append(index_set)
        random.seed(seed)
        random.shuffle(big_index_sets)
        random.shuffle(small_index_sets)
        index_sets = big_index_sets + small_index_sets
    else:  # Sort from largest to smallest doi sets
        index_sets = sorted(list(doi_to_indices.values()),
                            key=lambda index_set: len(index_set),
                            reverse=True)

    for index_set in index_sets:
        if len(train) + len(index_set) <= train_size:
            train += index_set
            train_doi_count += 1
        elif len(val) + len(index_set) <= val_size:
            val += index_set
            val_doi_count += 1
        else:
            test += index_set
            test_doi_count += 1

    
    print(f'Total dois = {len(doi_to_indices):,} | '
          f'train dois = {train_doi_count:,} | '
          f'val dois = {val_doi_count:,} | '
          f'test dois = {test_doi_count:,}')
    
    return train, val, test
```

```python
def doi_split_corrected(df, sizes=(0.8, 0.1, 0.1), seed=0):

    assert sum(sizes) == 1

    DOI = list(set(df.DOI)) # We extract the DOIs for each complexes
    length = len(df)

    # Split
    train, val, test = [], [], []
    train_doi_count, val_doi_count, test_doi_count = 0, 0, 0

    random.seed(seed)
    random.shuffle(DOI) # Randomly shuffle unique DOIs
    train_range = int(sizes[0] * len(DOI))
    val_range = int(sizes[1] * len(DOI))

    for i in range(train_range):
        condition = df['DOI'] == DOI[i] # We choose a unique DOI at random and select all permutations of the same complex with this ID
        selected = df[condition]
        indices = selected.index.tolist() # The absolute indices of these entries (which are consecutive in the dataframe) are added to the train set
        train_doi_count+=1
        for i in indices:
            train.append(i)

    for i in range(train_range, train_range + val_range):
        condition = df['DOI'] == DOI[i]
        selected = df[condition]
        indices = selected.index.tolist()
        val_doi_count+=1
        for i in indices:
            val.append(i)

    for i in range(train_range + val_range, len(DOI)):
        condition = df['DOI'] == DOI[i]
        selected = df[condition]
        indices = selected.index.tolist()
        test_doi_count+=1
        for i in indices:
            test.append(i)

    print(f'train dois = {train_doi_count:,} | '
          f'val dois = {val_doi_count:,} | '
          f'test dois = {test_doi_count:,}')

    return train, val, test
```

# DOI Splitting

We take the average of pIC50 for multpile occurences of a complexe (better than dropping duplicates).

```python
from sklearn.metrics.pairwise import rbf_kernel
rf = RandomForestRegressor(max_depth=13, n_estimators=310, min_samples_leaf=1, max_features=0.5, random_state=42)
```

```python
def cross_validation(indices, X, y, rf):
    y_data= []
    y_predictions = []

    for i, (train_idx, val_idx, test_idx) in enumerate(indices):
        print("CV iteration", i)
    
        rf.fit(X[train_idx], y[train_idx])   # Fit model to data
        y_pred = rf.predict(X[test_idx]) # Predict values
        y_data.extend(y[test_idx])
        y_predictions.extend(y_pred) #Update lists
    
    y_data = np.array(y_data)
    y_predictions = np.array(y_predictions)
    
    metrics = obtain_metrics(y_data, y_predictions)
    print(metrics)
    return y_data, y_predictions
```

```python
def get_indices(metals, CV):
    
    indices = []
    
    for seed in range(CV):
        random.seed(seed)
        train_indices, val_indices, test_indices = doi_split(metals,
                                                                  sizes=(0.9, 0, 0.1),
                                                                  balanced=True,
                                                                  seed=seed)
        print(len(train_indices), len(val_indices), len(test_indices))
        indices.append([train_indices, val_indices, test_indices])
    
    return indices
```

```python
def get_indices_corrected(metals, CV):
    
    indices = []
    
    for seed in range(CV):
        random.seed(seed)
        train_indices, val_indices, test_indices = doi_split_corrected(metals,
                                                                  sizes=(0.9, 0, 0.1),
                                                                  seed=seed)
        print(len(train_indices), len(val_indices), len(test_indices))
        indices.append([train_indices, val_indices, test_indices])
    
    return indices
```
