**NumAromaticHeterocycles** *(Feature #55)*: This descriptor counts the number of aromatic heterocycles present in a molecule. Aromatic heterocycles are ring structures containing at least one heteroatom (like nitrogen, oxygen, or sulfur) in addition to carbon atoms.

**SMR_VSA3** *(Feature #84)*: This descriptor is related to the Scaled Molecular Resonance VSA (Volume Steric Accessibility) descriptor. Specifically, it refers to the contribution of the molecular surface area associated with atoms that are bound to 3 heavy (non-hydrogen) atoms.

**fr_pyridine** *(Feature #187)*: This descriptor indicates the presence of the pyridine substructure in a molecule. Pyridine is a six-membered aromatic ring containing five carbon atoms and one nitrogen atom.

**VSA_EState2** *(Feature #106)*: This descriptor corresponds to the E-State (Electrotopological State) VSA (Volume Steric Accessibility) descriptor, focusing on the contribution of the molecular surface area associated with a specific atom type or hybridization state. In this case, it represents the contribution of atoms with an E-State value of 2.

**EState_VSA7** *(Feature #22)*: Similar to VSA_EState2, this descriptor represents the contribution of the molecular surface area associated with atoms having an E-State value of 7.

**VSA_EState3** *(Feature #107)*: Similar to VSA_EState2, but focusing on atoms with an E-State value of 3.

**SMR_VSA10** *(Feature #82)*: Similar to SMR_VSA3, but this descriptor focuses on the contribution of the molecular surface area associated with atoms bound to 10 heavy (non-hydrogen) atoms.

**VSA_EState7** *(Feature #111)*: Similar to EState_VSA7, this descriptor represents the contribution of the molecular surface area associated with atoms having an E-State value of 7.

**PEOE_VSA8** *(Feature #78)*: This descriptor is part of the PEOE (Partial Equalization of Orbital Electronegativity) type VSA descriptor. It represents the contribution of the molecular surface area associated with a specific atom type or hybridization state. In this case, it represents the contribution of atoms with a PEOE value of 8.

**MolLogP** *(Feature #46)*: This descriptor calculates the logarithm of the partition coefficient (logP) of the molecule. LogP is a measure of a molecule's hydrophobicity or lipophilicity, indicating its propensity to partition between hydrophobic and hydrophilic phases.

# What are the possible environments for each feature ? 

For ecxlusively the features with a positive impact on the pIC50, we look at the general environments of these features.

**For these following parts, we study the top features for the Morgan Fingerprints, but feel free to modify the code to see the results for RDKit Fingerprint.**

```python
def get_bit_smiles_df(df, top_features):
    
    all_smiles = list(metals['L1']) + list(metals['L2']) + list(metals['L3'])
    all_smiles_unique = list(set(all_smiles))

    #Bit_smiles is a Dataframe, with one row per top feature.
    #The BIT column contains the bit number of the feature.
    #The SMILES column contains a list of all the SMILES having the feature. 
    
    bit_smiles = pd.DataFrame(columns=['SMILES', 'BIT'])
    
    for bit in top_features:
        bit_smiles.loc[len(bit_smiles)] = [[], bit]
        
    for bit in top_features:
        for smiles in all_smiles_unique:
            info = {}
            fingerprint = rdMolDescriptors.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(smiles), radius=2, nBits=2048, bitInfo=info)
            if bit in info.keys():
                bit_smiles.loc[bit_smiles['BIT'] == bit, 'SMILES'] = bit_smiles.loc[bit_smiles['BIT'] == bit, 'SMILES'].apply(lambda x: x + [smiles])
            
    return bit_smiles
```

```python
# The function gethighlight_complexe is different than gethighlight, because it gets the ligand's number
# (1, 2 or 3) of the complexe we are looking at. 
def gethighlight_complexe(bit_smiles_ligand) :
    highlights = []
    highlights_texts = []
    
    for pair in bit_smiles_ligand: 
        smiles = pair[0]
        bit = pair[1]
        ligand_index = pair[2]
    
        mol = Chem.MolFromSmiles(smiles)
    
        info = {}
        rdMolDescriptors.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048, bitInfo=info)
        atomID, radius = info[bit][0]
  
        if radius>0:
            env = Chem.FindAtomEnvironmentOfRadiusN(mol, radius, atomID)
            atomsToUse=[]
            for b in env:
                atomsToUse.append(mol.GetBondWithIdx(b).GetBeginAtomIdx())
                atomsToUse.append(mol.GetBondWithIdx(b).GetEndAtomIdx())
            atomsToUse = list(set(atomsToUse))
        else:
            atomsToUse = [atomID]
        highlights.append(atomsToUse)
        highlights_texts.append('Ligand number' + str(ligand_index) + ' , Feature number' + str(bit))
    
    return highlights, highlights_texts
```

```python
def get_top_features(complexe, top_features): #complexe is a list of smiles
    
    bit_smiles_ligand = []
    features_count = 0

    for index, smiles in enumerate(complexe):
        info = {}
        fingerprint = rdMolDescriptors.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(smiles), radius=2, nBits=2048, bitInfo=info)
        for bit in top_features:
            if bit in info.keys():
                bit_smiles_ligand.append((smiles, bit, index))
                features_count +=1
    
    if features_count == 0 : 
        print('No top feature in this complexe')
        return False
    else : 
        highlights, highlights_text = gethighlight_complexe(bit_smiles_ligand)
        molecule_list = [Chem.MolFromSmiles(smiles[0]) for smiles in bit_smiles_ligand]
        img = Draw.MolsToGridImage(molecule_list, molsPerRow=5, subImgSize=(400, 400), legends=highlights_text, highlightAtomLists=highlights)
        return img
```

# Feature Importance Analysis

Substructures linked to top 10 features with the best pIC50. We run this code on the Morgan Fingerprint dataframe, but it can be adapted on any of the other df.

```python
rf = RandomForestRegressor()
X = np.array(metals_morgan['Fingerprint'].values.tolist())
y = np.array(metals_morgan.pIC50.values.tolist())
rf.fit(X, y)

# Getting feature importances
feature_importances = rf.feature_importances_

# Finding the indices of the 15 most important features
indices_top = np.argsort(feature_importances)[-15:]

print("Indices of 15 most important features:", indices_top)
```

```python
mol_smiles = []

for i in range(len(metals_morgan)):
    mol_smiles.append(f'{metals_morgan.L1.iloc[i]}.{metals_morgan.L2.iloc[i]}.{metals_morgan.L3.iloc[i]}')

metals_morgan['SMILES123'] = mol_smiles
```

```python
smiles_list = metals_morgan['SMILES123'].values.tolist()
mols_list = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]   # This seems a bit convoluted : why not define a new column with mol objects 
                                                                     # and then use 'for smiles in metals_morgan['mol']' ?
    
substructs = []
for mol in mols_list:
    bi = {}
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=512, bitInfo=bi)
    for bit in indices_top:
        if bit in bi.keys():
            
            # We calculate the fingerprint of each compound and verify whether or not it has one of the top features ; if so, we extract the substructure 
            # associated with the feature and add it to the list
            
            atom_idx, radius = bi[bit][0]
            env = Chem.FindAtomEnvironmentOfRadiusN(mol, radius, atom_idx)
            submol = Chem.MolToSmarts(Chem.PathToSubmol(mol, env))
            
            if submol not in substructs and submol != '':
                substructs.append(submol)
        else:
            continue
```

```python
# Draw substructures in grid
mols = [Chem.MolFromSmiles(substruct) for substruct in substructs]
Draw.MolsToGridImage(mols, molsPerRow=5, subImgSize=(200, 200), useSVG=True)
```
