```python
good_examples = good_examples.drop_duplicates(subset=['L3'])
```

```python
len(good_examples)
```

```python
B_ligands =list(set(good_examples['L3']))
```

```python
top6 = ['c1ccc(-c2ccccn2)nc1', 'c1cnc2c(c1)ccc1cccnc12', 'Cc1ccc2ccc3ccc(C)nc3c2n1', 'c1ccc(-c2ccnc3c2ccc2c(-c4ccccc4)ccnc23)cc1',
        'Cc1ccnc(-c2cc(C)ccn2)c1', 'c1ccc2nc3c4cccnc4c4ncccc4c3nc2c1', 'c1cc2c3c(c4cccnc4c2nc1)nc[nH]3']
A_lig = pd.DataFrame(columns=['SMILES'])
A_lig.SMILES = top6
A_lig['mol'] = A_lig['SMILES'].apply(Chem.MolFromSmiles)
mols2grid.display(A_lig, mol_col='mol')
```

```python
ligands_AAB=[]
for A in top6:
    for B in B_ligands:
        ligands_AAB.append({'L1' : A, 'L2' : A, 'L3' : B}) #two generic phen- or bipy-based ligands + 1 specific ligand
len(ligands_AAB)
```

```python
ligands_AAB = pd.DataFrame(ligands_AAB)
```

```python
ligands_AAB.to_csv('good_examples_synthesis.csv', index=False)
```
