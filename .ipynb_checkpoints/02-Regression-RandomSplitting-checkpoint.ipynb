### Data Preprocessing

```python
metals= pd.read_csv("ruthenium_complexes_dataset.csv", dtype={'L1': str, 'L2': str, 'L3': str})
metals.dropna(subset=['L1', 'L2', 'L3'], how='any', inplace=True)
metals.reset_index(drop=True, inplace=True)
```

```python
metals['MOL1'] = metals.L1.apply(Chem.MolFromSmiles)
metals['MOL2'] = metals.L2.apply(Chem.MolFromSmiles)
metals['MOL3'] = metals.L3.apply(Chem.MolFromSmiles)
metals['Ligands_Set'] = metals.apply(lambda row: set([row['L1'], row['L2'], row['L3']]), axis=1)

metals['ECFP4_1'] = metals.MOL1.apply(calc_morgan_fingerprint)
metals['ECFP4_2'] = metals.MOL2.apply(calc_morgan_fingerprint)
metals['ECFP4_3'] = metals.MOL3.apply(calc_morgan_fingerprint)

#metals['ECFP4_1'] = metals.MOL1.apply(calc_rdkit_fingerprint)
#metals['ECFP4_2'] = metals.MOL2.apply(calc_rdkit_fingerprint)
#metals['ECFP4_3'] = metals.MOL3.apply(calc_rdkit_fingerprint)
```

Clean the column names (remove any spaces, etc) and convert all IC50s to floats.

```python
metals.rename(columns={'IC50 (μM)': 'IC50', 'Incubation Time (hours)': 'IncubationTime', 'Partition Coef logP': 'logP', 'Cell Lines ': 'Cells'}, inplace=True)
```

Convert all IC50s to floats

```python
metals['IC50'] = metals['IC50'].apply(convert_to_float)
```

We will now add the fingerprints to generate a single fingerprint representing the 3 ligands. 

```python
add_lists = lambda row: [sum(x) for x in zip(row['ECFP4_1'], row['ECFP4_2'], row['ECFP4_3'])]
metals['Fingerprint'] = metals.apply(add_lists, axis=1)
```

```python
metals['pIC50'] = metals['IC50'].apply(lambda x: - np.log10(x * 10 ** (-6)))
```

## Model - Random Forest 

First, let's define some useful functions. 

```python
def perform_cross_validation_rf(df, n_splits, target, max_depth=13, n_estimators=310, min_samples_leaf=1, max_features=0.5):
    X = np.array(df['Fingerprint'].values.tolist())
    y = np.array(df[target].values.tolist())
    rf = RandomForestRegressor(max_depth=max_depth, n_estimators=n_estimators, min_samples_leaf=min_samples_leaf, max_features=max_features, random_state=42)
    cv = KFold(n_splits=n_splits, shuffle=True, random_state=42)

    # Generate the predictions
    y_data, y_predictions = run_cross_validation(X, y, rf, cv)

    metrics = obtain_metrics(y_data, y_predictions)
    print(metrics)

    return y_data, y_predictions

# General function to run cross validation
def run_cross_validation(X, y, model, cv):
    y_data = []
    y_predictions = []
    for train, test in cv.split(X, y):
        model.fit(X[train], y[train])   # Fit model to data
        y_pred = model.predict(X[test]) # Predict values
        y_data.extend(y[test])
        y_predictions.extend(y_pred) #Update lists
    y_data = np.array(y_data)
    y_predictions = np.array(y_predictions)   #Turn the lists into arrays 

    return y_data, y_predictions
```

## Influence of the training size

We can define a function that plots the learning curve of our estimator. Here we iterate on different scoring metrics ; since MSE and MAE are computed as negative numbers we change signs for the score arrays when we ask for such metrics.

```python
def learning_curves_cv_rf(df, n_splits, target, samples, metrics=['neg_mean_absolute_error'], max_depth=9, n_estimators=20, min_samples_leaf=4, max_features=0.5):
    X = np.array(df['Fingerprint'].values.tolist())
    y = np.array(df[target].values.tolist())
    rf = RandomForestRegressor(max_depth=max_depth, n_estimators=n_estimators, min_samples_leaf=min_samples_leaf, max_features=max_features, random_state=42)
    cv = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    i=0
    
    fig, axs = plt.subplots(1, 3, figsize=(25, 8))
    
    for scoring in metrics :
        
        # Implementing the learning curve
        train_sizes, train_scores, test_scores = learning_curve(rf, X, y, cv=cv, train_sizes=samples, scoring=scoring)
    
        # Adressing the sign issue with MSE, RMSE, MAE, etc
        if scoring.startswith('neg'):
            negate_score=True
        else:
            negate_score=False
    
        display = LearningCurveDisplay(train_sizes=train_sizes, train_scores=train_scores, test_scores=test_scores, score_name="Score")

        display.plot(ax=axs[i], score_type='both', negate_score=negate_score)
        title = scoring.replace('neg','')
        title = title.replace('_',' ')
        axs[i].set_title(title)
        axs[i].set_xlabel('Size of the training sample', fontsize=15)
        axs[i].set_ylabel('Score', fontsize=15)
        plt.gca().legend(['Train', 'Test'], loc='best') # Customize legend labels

        i+=1

    plt.show()
```

First we use the whole dataset : we re-use the metals_train1 dataframe defined above. We first target IC50 values.

```python
metrics = ['neg_mean_absolute_error', 'neg_root_mean_squared_error', 'r2']

learning_curves_cv_rf(metals_train1, 10, 'IC50', samples=np.linspace(0.1, 1, 10), metrics=metrics)
```

We can do the same for pIC50 values :

```python
learning_curves_cv_rf(metals_train1, 10, 'pIC50', samples=np.linspace(0.1, 1, 10), metrics=metrics)
```

## Handling the outliers
