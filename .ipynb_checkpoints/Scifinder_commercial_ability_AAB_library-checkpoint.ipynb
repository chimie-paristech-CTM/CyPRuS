```python
from random import sample, seed
import numpy as np
import pandas as pd
#!pip install rdkit mols2grid
from rdkit import Chem, RDLogger
from rdkit.Chem import AllChem, DataStructs, Draw, rdMolDescriptors

# Visualization
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})

import mols2grid

import os

from tqdm.auto import tqdm
import time


np.random.seed(42)
seed(42)
```

```python
directory = "../Scifinder/nitrogen"

file_list = os.listdir(directory)

df_list=[]

for file in file_list:
    file_path = os.path.join(directory, file)
    df=pd.read_csv(file_path, on_bad_lines='warn', encoding='unicode_escape')
    for i in range(len(df['Canonical SMILES'])):
        smiles=df['Canonical SMILES'][i]
        df.at[i, 'Canonical SMILES']=str(smiles)
    filtered_df = df[~df['Canonical SMILES'].str.contains('\.')]
    df_list.append(filtered_df)

scifinder_nitrogen_ligands = pd.concat(df_list, ignore_index=True)
```

```python
scifinder_nitrogen_ligands
```

```python
scifinder_nitrogen_ligands['SMILES'] = scifinder_nitrogen_ligands['Canonical SMILES']
```

```python
for i in range(len(scifinder_nitrogen_ligands.SMILES)):
    if scifinder_nitrogen_ligands.SMILES[i] == 'nan':
        scifinder_nitrogen_ligands.at[i, 'SMILES']=None
    if not scifinder_nitrogen_ligands.SMILES[i]:
        scifinder_nitrogen_ligands.at[i, 'SMILES']=scifinder_nitrogen_ligands['Isomeric SMILES'][i]
```

```python
scifinder_nitrogen_ligands['SMILES']
```

```python
scifinder_nitrogen_ligands.dropna(subset=['SMILES'], how='any', inplace=True)
scifinder_nitrogen_ligands.reset_index(drop=True, inplace=True)
```

```python
scifinder_nitrogen_ligands
```

```python
directory = "../Scifinder/specific"

file_list = os.listdir(directory)

df_list=[]

for file in file_list:
    file_path = os.path.join(directory, file)
    df=pd.read_csv(file_path, on_bad_lines='warn', encoding='unicode_escape')
    for i in range(len(df['Canonical SMILES'])):
        smiles=df['Canonical SMILES'][i]
        df.at[i, 'Canonical SMILES']=str(smiles)
    filtered_df = df[~df['Canonical SMILES'].str.contains('\.')]
    df_list.append(filtered_df)

scifinder_specific_ligands = pd.concat(df_list, ignore_index=True)
```

```python
scifinder_specific_ligands['SMILES'] = scifinder_specific_ligands['Canonical SMILES']

for i in range(len(scifinder_nitrogen_ligands.SMILES)):
    if scifinder_specific_ligands.SMILES[i] == 'nan':
        scifinder_specific_ligands.at[i, 'SMILES']=None
    if not scifinder_specific_ligands.SMILES[i]:
        scifinder_specific_ligands.at[i, 'SMILES']=scifinder_specific_ligands['Isomeric SMILES'][i]
```

```python
scifinder_specific_ligands.dropna(subset=['SMILES'], how='any', inplace=True)
scifinder_specific_ligands.reset_index(drop=True, inplace=True)
```

```python
scifinder_specific_ligands
```

```python
phen= Chem.MolFromSmarts('c1[cH][n]c2c(c1)ccc1cc[cH][n]c12')
bipy = Chem.MolFromSmarts('c1cc[cH][n]c1-c2[n][cH]ccc2')
dimebipy = Chem.MolFromSmarts('c1ccc([CH3])[n]c1-c2[n]c([CH3])ccc2')
dimephen= Chem.MolFromSmarts('c1c([CH3])[n]c2c(c1)ccc1ccc([CH3])[n]c12')
```

```python
# We filter nitrogen-based ligands as we don't want substituents other than methyl on ortho positions of bipy and phen

nitrogen = []
for smiles in scifinder_nitrogen_ligands['SMILES'] :
    smiles=str(smiles)
    mol=Chem.MolFromSmiles(smiles)
    if mol:
        if mol.HasSubstructMatch(bipy) or mol.HasSubstructMatch(phen) or mol.HasSubstructMatch(dimebipy) or mol.HasSubstructMatch(dimephen):
            nitrogen.append({'SMILES' : smiles, 'mol' : mol})

nitrogen_ligands=pd.DataFrame(nitrogen)
#nitrogen_ligands.to_csv('scifinder_bipy_phen_ligands_no_sub_ortho.csv')
```

```python
quinoleine = Chem.MolFromSmarts('c1cc2c(c([OH])[cH]1)[n][cH]cc2')
dimequinoleine = Chem.MolFromSmarts('c1cc2c(c([OH])c([CH3])1)[n]c([CH3])cc2')
dipyamine = Chem.MolFromSmarts('c1ccc(Nc2ccc[cH]n2)n[cH]1')
dimedipyamine = Chem.MolFromSmarts('c1ccc(Nc2cccc([CH3])n2)nc([CH3])1')
framework13 = Chem.MolFromSmarts('O=c1c2[cH]cccc2nc(-c2ccccc2)n1/N=C/c1ccccc1')
meframework13 = Chem.MolFromSmarts('O=c1c2c([CH3])cccc2nc(-c2ccccc2)n1/N=C/c1ccccc1')
```

```python
# We apply similar filtering rules to specific ligands

RDLogger.DisableLog('rdApp.*')

specific = []
for smiles in scifinder_specific_ligands.SMILES :
    smiles=str(smiles)
    mol=Chem.MolFromSmiles(smiles)
    if mol:
        if mol.HasSubstructMatch(quinoleine) or mol.HasSubstructMatch(dipyamine) or mol.HasSubstructMatch(framework13) or mol.HasSubstructMatch(dimequinoleine) or mol.HasSubstructMatch(dimedipyamine) or mol.HasSubstructMatch(meframework13):
            specific.append({'SMILES' : smiles, 'mol' : mol})

specific_ligands=pd.DataFrame(specific)
```

```python
len(specific), len(nitrogen)
```

```python
AAB=[]
for A in nitrogen_ligands.SMILES:
    for B in specific_ligands.SMILES:
        AAB.append({'L1' : A, 'L2' : A, 'L3' : B}) #two generic phen- or bipy-based ligands + 1 specific ligand
    for C in nitrogen_ligands.SMILES:
        AAB.append({'L1' : A, 'L2' : A, 'L3' : C}) #let's not forget that the specific ligand can be phen- or bipy-based
```

```python
AAB_scifinder_compounds=pd.DataFrame(AAB)
```

## Predictions over the library

```python
prediction=pd.read_csv('Prediction_63k_scifinder.csv')
```

```python
pIC50=prediction.pIC50.tolist()
```

```python
variance=prediction.Vars.tolist()
```

```python
bins = np.arange(3, 7.1, 0.1)
fig, axes = plt.subplots(1, 2, figsize=(18, 5))  # 1 row, 2 columns

# Plot histograms for each series
axes[0].hist(pIC50, bins=bins, color='blue', alpha=0.5)
axes[1].hist(variance, bins=bins, color='green', alpha=0.5)

# Customize plot labels and ticks
axes[0].set_xlabel('pIC50')
axes[0].set_ylabel('Occurrences')
axes[0].set_title('pIC50 predictions repartition')
axes[0].set_xticks(bins)  # Set x-axis ticks to bin edges

axes[0].tick_params(axis='x', labelsize=6)

axes[1].set_xlabel('Deviation')
axes[1].set_ylabel('Occurrences')
axes[1].set_title('pIC50 predictions deviations')
axes[1].set_xticks(bins)  # Set x-axis ticks to bin edges

axes[1].tick_params(axis='x', labelsize=6)
# Adjust layout
#plt.tight_layout()

# Show the plot
plt.show()
```
