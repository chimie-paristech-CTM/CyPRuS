```python
def get_ligands_dict(row, global_dict):
    """
    Stores unique ligands from the L1, L2 and L3 columns in a row in a global dictionary.
    
    Args:
        row: a row of a dataframe
        global_dict: a dictionary to update with the mol3D instance of each unique SMILES string
    
    Returns:
        None
    """
    for col in ['L1', 'L2', 'L3']:
        string = row[col]
        if string not in global_dict:
            global_dict[string] = get_mol3D(string)

def create_ligands_dict(df):
    """
    Creates a dictionary of all different ligands in a dataframe made up of columns L1, L2, L3.
    
    Args:
        df: a pandas DataFrame containing columns L1, L2, L3
    
    Returns:
        ligands_dict (dict): a dictionary containing the SMILES strings with their corresponding mol3D instance
    """
    ligands_dict = {}
    df.apply(lambda row: get_ligands_dict(row, ligands_dict), axis=1)
    return ligands_dict
```

```python
def get_mol3D_df(row, ligands_dict, i):
    """
    Looks up the mol3D object corresponding to the SMILES string in the ligands_dict based on the column index.
    
    Args:
        row: a row of a dataframe
        ligands_dict: a dictionary containing SMILES strings with their corresponding mol3D instances
        i: integer indicating the column index (1, 2, or 3)
    
    Returns:
        mol_instance: mol3D instance corresponding to the SMILES string in the specified column
    """
    mol = mol3D()
    column = f'L{i}'
    mol.copymol3D(ligands_dict.get(row[column], None))
    return 
```

```python
def sanitize_filename(filename):
    
    """
    Replace invalid characters in a filename with underscores and remove consecutive dashes.

    Parameters:
        filename (str): The filename to sanitize.

    Returns:
        str: The sanitized filename.
    """
    
    invalid_chars = {'/', '\\', ':', '*', '?', '"', '<', '>', '|', '@'}   
    sanitized_filename = ''.join(c if c not in invalid_chars else '-' for c in filename)
    sanitized_filename = sanitized_filename.replace('--', '-')
    
    return sanitized_filename


def max_filename_length(folder_path):
    
    """
    Get the maximum allowed length of filenames in the specified folder path.

    Parameters:
        folder_path (str): The path to the folder for which to determine the maximum filename length.

    Returns:
        int: The maximum allowed length of filenames in the specified folder path.
    """
    
    max_length = os.pathconf(folder_path, 'PC_NAME_MAX')

    return max_length


def get_mol_file_name(smiles, folder_path):
    """
    Generate a valid filename for a .mol file based on the given SMILES string.

    Parameters:
        smiles (str): The SMILES string.
        folder_path (str): The path to the folder where the .mol file will be saved.

    Returns:
        str: The sanitized and truncated filename for the .mol file.
    """
    max_length = max_filename_length(folder_path)
    filename = sanitize_filename(smiles)
    filename_with_extension = filename + "_truncated_name.mol"

    if len(filename_with_extension) > max_length:
        max_length_without_extension = max_length - len("_truncated_name.mol") - 1
        truncated_filename = filename[:max_length_without_extension]
        truncated_filename_with_extension = truncated_filename + "_truncated_name.mol"
    else:
        truncated_filename_with_extension = filename_with_extension
    
    return truncated_filename_with_extension


def create_mol_files(smiles_dict, folder_path):
    """
    Create .mol files from a dictionary of SMILES strings.

    Parameters:
        smiles_dict (dict): Dictionary with SMILES strings as keys.
        folder_path (str): Path to the folder where .mol files will be saved.

    Returns:
        None
    """
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    for smiles, _ in smiles_dict.items():
        mol = Chem.MolFromSmiles(smiles)
        mol_file_name = os.path.join(folder_path, get_mol_file_name(smiles, folder_path))
        Chem.MolToMolFile(mol, mol_file_name)
```

### Second step : updating the ligands.dict

```python
ligands_dict = {}

with open('molSimplify/molSimplify/Ligands/ligands.dict', 'r') as f:
    for line in f:
        if line.startswith('#') or not line.strip():
            continue
        
        key, value = line.strip().split(':', 1)
        attributes = value.split(',')
        
        ligands_dict[key] = {
            'structure_file': attributes[0],
            'abbreviation': attributes[1],
            'connection_atoms': attributes[2],
            'groups': attributes[3],
            'force_field': attributes[4],
            'formal_charge': attributes[5]
        }

print(ligands_dict)
```

```python
def calculate_charge_from_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None  # Unable to parse SMILES
    try:
        charge = AllChem.GetFormalCharge(mol)
        return charge
    except Exception as e:
        return None
```

```python
def update_ligands_dict(ligands_dict, ligands_df, folder_path):
    """
    Update the ligands_dict with new entries based on ligands_df.
    
    Parameters:
        ligands_dict (dict): The original dictionary of ligands.
        ligands_df (dict): Dictionary containing new ligand information.
        folder_path (str): The path to the folder where .mol files will be saved.
    
    Returns:
        dict: The updated dictionary of ligands.
    """
    for key in ligands_df:
        smiles = key  # Assuming the keys of ligands_df contain the SMILES strings
        structure_file = get_mol_file_name(smiles, folder_path)
        charge = calculate_charge_from_smiles(smiles)
        
        ligands_dict[key] = {
            'structure_file': structure_file,
            'abbreviation': key,
            'connection_atoms': None,
            'groups': 'build bidentate',
            'force_field': 'BA',
            'formal_charge': charge
        }
    
    return ligands_dict
```

```python
ligands_dict = update_ligands_dict(ligands_dict, ligands_df, "molSimplify/molSimplify/Ligands")
```

#### Let's check if our truncation method for the files names risk a mismatch. 

```python
def check_duplicates(ligands_dict, attribute):
    """
    Check for duplicates in the specified attribute values of ligands_dict and print them.
    
    Parameters:
        ligands_dict (dict): The dictionary of ligands.
        attribute (str): The attribute to check for duplicates.
    
    Returns:
        dict: A dictionary where the keys are the duplicate attribute values
              and the values are lists of keys that share the same attribute value.
    """
    attribute_values = {}
    duplicates = {}

    for key, value in ligands_dict.items():
        attribute_value = value.get(attribute)
        if attribute_value in attribute_values:
            if attribute_value not in duplicates:
                duplicates[attribute_value] = [attribute_values[attribute_value]]
            duplicates[attribute_value].append(key)
        else:
            attribute_values[attribute_value] = key

    if duplicates:
        print("Duplicate values found for attribute '{}':".format(attribute))
        for attr_value, keys in duplicates.items():
            print("Value '{}': {}\n".format(attr_value, ', '.join(keys)))
    else:
        print("No duplicates found for attribute '{}'.".format(attribute))

    return duplicates
```

```python
duplicates = check_duplicates(ligands_dict, 'structure_file')
```

The only duplicates in file names are previous ligands. So our truncation method is ok.

```python
def filter_dict_smarts(smiles_dict, smarts_list):
    selection = []
    not_selected = []

    for key in smiles_dict.keys():
        smiles = str(key)
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            if any(mol.HasSubstructMatch(sub) for sub in smarts_list):
                smiles = Chem.MolToSmiles(mol)
                mol = Chem.MolFromSmiles(smiles)
                selection.append({'Key': key, 'SMILES': smiles, 'mol': mol})
            else:
                not_selected.append({'Key': key, 'SMILES': smiles, 'mol': mol})

    selected_df = pd.DataFrame(selection)
    not_selected_df = pd.DataFrame(not_selected)

    return selected_df, not_selected_df
```
